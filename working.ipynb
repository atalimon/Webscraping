```python
# Importing libraries for web scraping and data manipulation.
# We use pandas to store and manipulate data, its like EXCEL or Gspreadsheets for python!
import pandas as pd 
# We use beautifulsoup to scrape data from website.
from bs4 import BeautifulSoup
# Requests is for make an request to target destination which is tmcars.com
import requests
# Regex is for string exploration and manipulation!
import re
# Numpy is also an data manipulation library
import numpy as np
```

```python
# Creating object for URL
tmcars_url = 'https://tmcars.info/tm/cars/all?brandId=12&modelIds=15&years=2021'
# Using BS4 requests method to GET data from TmCARS.
# Creating new object to save what we got from our request from TMCARS.
page_camry21 = requests.get(tmcars_url)
# Using soup method to parse, data. 
soup = BeautifulSoup(page_camry21.text, 'html.parser')
```

```python
# We creating new lists to keep our data, in this case its our links. 
profile_links = []
linko =[]
```

```python
# creating a function to scrape profile links from tmcars.
def createlinks():
    a_tag = soup.find_all('a', href = re.compile('^https://tmcars.info/tm/cars/'))
    for tag in a_tag:
        profile_links.append(tag['href'])
    mash = set(profile_links)
    for i in mash:
        linko.append(i)
```

```python
clear = []
for links in linko:
    if len(links) == 56:
        clear.append(links)
```

```python
# Creating Dataframe to store our scraped data.
df = pd.DataFrame()
# Using header to escape problems with too many requests.
header = {'User-Agent': 'python-requests/2.28.1', 'Accept-Encoding': 'gzip, deflate', 'Accept': '*/*', 'Connection': 'keep-alive'}
# Using foor loop to get data and store each to to the list. 
for s in range(len(clear)):
    sponse = requests.get(clear[s],header)
    src = sponse.text
    sup = BeautifulSoup(sponse.text, 'html.parser')
# First we find table from data.
    card = sup.find('table')
    if card not in sup:
        pass
# We find exact place where our data stores. 'tr' means for rows in html 'td' for columns.
    cardp = sup.find('div', class_='arrow-ribbon2 bg-primary')
    price = [k.text for k in cardp]
    cardlisto = [td.text  for td in card.findChildren('td')]
# I used while loop for check if profile contains each data we want. Exact amount is 14
# If data is not full our loop will add NA to our list    
    while len(cardlisto) != 13:
        cardlisto.append('NA')
    joined = price + cardlisto
    df = df.append(pd.DataFrame([joined], columns=['Price','Brand','Yyly', 'Yorediji', 'Motor', 'Gearbox', 'Probeg', 'Renk', 'Vin','Kredit', 'Obmen', 'Location', 'Telno', '14']),ignore_index=True)
```

```python
df['Price'] = df['Price'].str.replace(r'\D', '')
df['Probeg'] = df['Probeg'].str.replace(r'\D', '')
df['Motor'] = df['Motor'].str.replace(r'\D', '')
```
